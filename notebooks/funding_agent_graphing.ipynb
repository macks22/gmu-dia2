```python
import funding_agents
dirs = funding_agents.dirs()
fig, ax = plt.subplots()
ax.set_ylabel('Number of Awards Funded')
dirs.plot(ax=ax, kind='bar', x='dir_abbr', y='freq', color='g', title='Directorates by Number of Awards Funded')
ax.set_xlabel('Directorates')
ax.grid(False)
fig = plt.gcf()
fig.set_size_inches(14, 8)
savefig('dirs-by-num-awards-funded.png')
```

```python
selector = funding_agents.DF['dir_id'] == '05'
cols = ['div_id', 'div_abbr', 'div_name']
divs = funding_agents.DF[selector][cols]
divs['freq'] = divs.groupby('div_id')['div_id'].transform('count')
divs = divs.sort('freq', ascending=False)
divs = divs.drop_duplicates().reset_index(drop=True)
divs

fig, ax = plt.subplots()
ax.set_ylabel('Number of Awards Funded')
divs.plot(ax=ax, kind='bar', x='div_abbr', y='freq', color='g', title='CISE Divisions by Number of Awards Funded')
ax.set_xlabel('Divisions')
ax.grid(False)
fig = plt.gcf()
fig.set_size_inches(14, 8)
savefig('cise-divs-by-num-awards-funded.png')
```

```python
print len(divs)
divs
```

```python
selector = funding_agents.DF['dir_id'] == '05'
cols = ['div_id', 'pgm_id', 'pgm_name']
pgms = funding_agents.DF[selector][cols]
pgms['freq'] = pgms.groupby('pgm_id')['pgm_id'].transform('count')
pgms = pgms.sort('freq', ascending=False)
pgms = pgms.drop_duplicates().reset_index(drop=True)
pgms

fig, ax = plt.subplots()
ax.set_ylabel('Number of Awards Funded')
pgms.head(40).plot(kind='bar', y='freq', color='g', title='CISE Programs by Number of Awards Funded')
ax.set_xlabel('Programs')
ax.set_xticklabels(pgms.head(40)['pgm_id'], rotation=45)
ax.grid(False)
fig = plt.gcf()
fig.set_size_inches(20, 10)
savefig('cise-pgms-by-num-awards-funded.png')
```
